# Met data

This notebook serves the sole purpose of plotting meteorological data on the period of interest for our work with the lidar (Dec. 2021 - Feb. 2022). The objective is to be able to print this data and manually reconstruct the history of the snowpack.

```python
import csv
import matplotlib.pyplot as plt
```

```python
time_series = []
time_converted_series = []
fc1_drift_series = []
fc2_drift_series = []
wind_speed_series = []
wind_direction_series = []
air_temp_series = []
surf_temp_series = []
precip_series = []
total_precip_series = []

with open('/home/mabonnet/Desktop/data/202202_finse_livox/met_obs.csv') as met_data:
    reader = csv.reader(met_data)
    header = next(reader)
    
    for row in reader:
        time_series.append(row[0])
        time_converted_series.append(float(row[1]))
        
        if row[2]!='':
            fc1_drift_series.append(float(row[2]))
        else:
            fc1_drift_series.append(None)
            
        if row[3]!='':
            fc2_drift_series.append(float(row[3]))
        else:
            fc2_drift_series.append(None)
        
        if row[5]!='':
            wind_speed_series.append(float(row[5]))
        else:
            wind_speed_series.append(None)
            
        if row[4]!='':
            wind_direction_series.append(float(row[4]))
        else:
            wind_direction_series.append(None)
            
        if row[6]!='':
            air_temp_series.append(float(row[6]))
        else:
            air_temp_series.append(None)
            
        if row[8]!='':
            surf_temp_series.append(float(row[8]))
        else:
            surf_temp_series.append(None)
            
        if row[7]!='':
            precip_series.append(float(row[7]))
        else:
            precip_series.append(0)
            
        if row[10]!='':
            total_precip_series.append(float(row[10]))
        else:
            total_precip_series.append(None)
    
    
    
    index_dec = time_series.index('2021-12-01 00:00:00')
    wind_direction_series = wind_direction_series
    
    time_series = time_series[index_dec:]
    time_converted_series = time_converted_series[index_dec:]
    fc1_drift_series = fc1_drift_series[index_dec:]
    fc2_drift_series = fc2_drift_series[index_dec:]
    wind_speed_series = wind_speed_series[index_dec:]
    wind_direction_series = wind_direction_series[index_dec:]
    air_temp_series = air_temp_series[index_dec:]
    surf_temp_series = surf_temp_series[index_dec:]
    precip_series = precip_series[index_dec:]
    total_precip_series = total_precip_series[index_dec:]
```

```python
plt.figure(figsize=(28, 7))
plt.plot(time_converted_series, fc1_drift_series, marker='.', label='fc1 drift')
plt.plot(time_converted_series, fc2_drift_series, marker='.', label='fc2 drift')
plt.legend()
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel('Time units')
plt.ylabel('fc units')
#plt.savefig("met_fig_fc.png")
```

```python
plt.figure(figsize=(28, 4))
plt.plot(time_converted_series, wind_speed_series, marker='.')
plt.xlabel('Time units')
plt.ylabel('Wind speed')
#plt.savefig("met_fig_ws.png")
```

```python
plt.figure(figsize=(28, 8))
plt.plot(time_converted_series, air_temp_series, linewidth=2, label='air temperature')
plt.plot(time_converted_series, surf_temp_series, linewidth=0.8, label='surface temperature')
plt.axhline(y=0, color='r', linestyle='--', label='0 degree line')
plt.legend()
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel('Time units')
plt.ylabel('Temperature')
#plt.savefig("met_fig_temp.png")
```
